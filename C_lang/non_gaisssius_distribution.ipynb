```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define parameters of the pareto distibution
alpha = 3
xm = 1

#create an array of x  values
x = np.linspace(0.1,10,1000)


#claculataed the y values of the patero distribution

y = alpha * (xm**alpha)/(x**(alpha+1))

plt.plot(x,y)
plt.show()

```

```python
plt.plot(np.log(x),np.log(y))

#add labels and a title
plt.xlabel("x")
plt.ylabel("p(x)")
plt.title("pareto distribution")

plt.show()
```
